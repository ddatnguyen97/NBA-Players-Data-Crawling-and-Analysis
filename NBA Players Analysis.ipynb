```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('NBA Players Season 23 - 24 Stats.csv')
data.head(20)
```

```python
point_per_game = (data['PTS']/data['GP']).round(1)
point_per_game
```

```python
data['PPG'] = point_per_game
```

```python
data.info()
```

```python
data = data.drop(data.columns[0], axis=1)
data.head()
```

```python
stats_by_team = data.groupby('TEAM').mean()
stats_by_team
```

```python
points_by_team = stats_by_team['PTS'].sort_values(ascending=False)
print(points_by_team)
figure = plt.figure(figsize=(15, 6))
sns.barplot(x=points_by_team, y=points_by_team.index)
plt.title('Average Points Scored by Team')
```

```python
top_10_3ppg = data.sort_values('PPG', ascending=False, ignore_index=True).head(10)
top_10_3ppg
```

```python
figure = plt.figure(figsize=(15, 6))
sns.scatterplot(x='PPG', y='FG%', data=data, size="3P%", hue="TEAM", sizes=(20, 400))
plt.title('Points per Game vs Field Goal Percentage by Player')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=5)

for line in range(0, top_10_3ppg.shape[0]):
    plt.text(top_10_3ppg['PPG'].iloc[line], 
             top_10_3ppg['FG%'].iloc[line], 
             top_10_3ppg['PLAYER'].iloc[line], 
             verticalalignment= 'bottom', 
             horizontalalignment= 'left', 
             size='small', color='black', 
             weight='semibold')
```
